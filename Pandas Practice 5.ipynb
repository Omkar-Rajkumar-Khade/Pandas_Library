# Data Selection & Processing

## loc() method

```python
import pandas as pd
```

```python
data2 = pd.read_csv('D:\\Datasets\\Test Data\\Employee_Dataset_3.csv')
print (data2) 
```

```python
#Find out values of particular index
x=data2.loc[2]
y=data2.loc[10]
print(x,'\n \nValue of another index \n\n',y)
```

```python
#Access values of multiple indexes simultaneously
data2.loc[[2,20,23]]
```

```python
#get value from particular column
data2.loc[6,'Department']
```

```python
#Slicing
data2.loc[2:20]
```

```python
data2.loc[5:15:2]
```

```python
#Slicing for particular column
data2.loc[6:12,'Department']
```

```python
#Conditional Operations For particular column
data2.loc[data2['Age']<30]
```

```python
data2.loc[data2['Age']<30,['Department']]
```

# Grouping Of Data

```python
data2
```

```python
#creating Group for Education column
gr1=data2.groupby(by='Education')
gr1
```

```python
gr1.groups
```

```python
#Access Group in List format
list(gr1)
```

```python
#Access Group in Dictionary format
dict(list(gr1))
```

```python
#creating Group for Mutiple columns
gr2=data2.groupby(['Education','Age'])
print(gr2)
gr2.groups
```

```python
dict(list(gr2))
```

```python
#use of get_group()
gr4=data2.groupby('Department').get_group('Software Development')
gr4
```
