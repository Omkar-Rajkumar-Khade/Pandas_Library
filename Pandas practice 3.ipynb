# Handling Missing Values in data

```python
df = pd.read_csv('D:\\Datasets\\Test Data\\Employee_Dataset_2.csv')
print (df)
```

```python
#converting particular string to Nan value
df1 = pd.read_csv('D:\\Datasets\\Test Data\\Employee_Dataset_2.csv',na_values=['not found','no values'])
print (df1)
```

```python
#converting particular string to Nan value(Column-Wise)
df2 = pd.read_csv('D:\\Datasets\\Test Data\\Employee_Dataset_2.csv',na_values={'Department':'no values','Marital Status':'not found'})
print (df2)
```

```python
#we dont want Nan in our data, we want our original values
df3 = pd.read_csv('D:\\Datasets\\Test Data\\Employee_Dataset_2.csv',keep_default_na=False)
print (df3)
```

```python
#check all dataset and return  
df.isnull()
```

```python
#Print total and column wise null values
column_null=df.isnull().sum()
total_null=df.isnull().sum().sum()
print("Total null values:",total_null)
print("Columnwise Null Values:\n")
print(column_null)
```

```python
#notnull is just opposite of isnull
df.notnull()
```

```python
#drop entire rows which has empty/null values
df.dropna()
```

```python
#drop entire columns which has empty/null values
df.dropna(axis=1)
```

```python
#remove particular rows by giving column name 
df.dropna(subset=['Department','Age'])
```
