# Filling Empty Values In Data

## Fillna() Method

```python
import pandas as pd
```

```python
data = pd.read_csv('D:\\Datasets\\Test Data\\Employee_Dataset_3.csv')
print (data) 
```

```python
#replace all empty values with some scalar number
data.fillna(5)
```

```python
#replace all empty values with some string value
data.fillna('data')
#some gets applied for other data types eg. float,char,int etc
```

```python
#column wise value filling 
data.fillna({'Age':30,'Department':'Data','Marital Status':'data2','Education':'Data3','Monthly Income':3000})
```

```python
#fill backward row data into empty valued row
data.fillna(method='bfill')
```

```python
#fill forward row data into empty valued row
data.fillna(method='ffill')
```

```python
#fill backward column data into empty valued column
data.fillna(method='bfill',axis=1)
```

```python
#fill forward column data into empty valued column
data.fillna(method='ffill',axis=1)
```

## Replace() Method

```python
data2 = pd.read_csv('D:\\Datasets\\Test Data\\Employee_Dataset_3.csv')
print (data2) 
```

```python
#Replace 'no values' with 'Data 3'
data2.replace('no values','Data 3')

# OR 
# data2.replace(to_replace='no values', value='Data 3')
```

```python
#Replace 35 with 50
data2.replace(35,50)
```

```python
#Replace Mutiple values
data2.replace(['Masters','Doctorate','Undergraduate',35],['Data 3','Data 4','Data 5',60])
```

```python
#Replace Column Wise
data2.replace({'Department':'Software Development','Education':'Undergraduate'},'IT Services')
```

```python
#relace all string values into integer
data2.replace('[A-Za-z]',0, regex=True)
```

```python
#relace all string values into integer for particular column
data2.replace({'Department':'[A-Za-z]'},0, regex=True)
```
