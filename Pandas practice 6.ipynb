# Merging of Dataframes/Datasets

## Merge() Function

```python
import pandas as pd
```

```python
#Create 1st Dataframe
df1=pd.DataFrame({'ID':[1,2,3,4,5],'class':[10,11,12,13,14,]})
df1
```

```python
#Create 2nd Dataframe
df2=pd.DataFrame({'ID':[1,2,3,4,5],'Name':['A','B','C','D','E']})
df2
```

```python
#merge both dataframes
pd.merge(df1,df2,on='ID')
```

```python
pd.merge(df2,df1 ,on='ID')
```

```python
df3=pd.DataFrame({'ID':[1,2,3,4,5,6],'Name':['A','B','C','D','E','G']})
df3
```

```python
#when we merge 2 different size dataframes then only commomn part gets merge
pd.merge(df1,df3)
```

```python
#Merge all elements of both dataframe
pd.merge(df1,df3,on='ID',how='outer')
```

```python
#Use of Indicator method
pd.merge(df1,df3,on='ID',how='outer',indicator=True)
```

```python
df4=pd.DataFrame({'ID':[5,6,8,41,45,6],'Name':['A','B','C','D','E','G']})
```

```python
#When you Have 2 different dataframes and different indexes then
pd.merge(df4,df3,left_index=True,right_index=True)
```

## Concat() Function

```python
#concatination for series
sr1=pd.Series([0,2,3,5,6])
sr1
```

```python
sr2=pd.Series([5,9,6,7,1,8])
sr2
```

```python
sr3=pd.Series([10,9,6,7,10,8])
sr3
```

```python
#concating 2 series
pd.concat([sr1,sr2])
```

```python
pd.concat([sr1,sr2],axis=1)
```

```python
pd.concat([sr1,sr2,sr3],axis=1)
```

```python
#concainnation for dataframes
pd.concat([df1,df2])
```

```python
#concainnation for dataframes column wise
pd.concat([df1,df2],axis=1)
```

```python
df5=pd.DataFrame({'ID':[5,6,8,],'Name':['A','B','C']})
df5
```

```python
#concate 2 unequal dictionaries
pd.concat([df4,df5],axis=1)
```

```python
#use of join function
a=pd.concat([df4,df5],axis=1,join='outer')
b=pd.concat([df4,df5],axis=1,join='inner')
print(a)
print('\n\n')
print(b)
```

```python
#we use keys to label 2 dataframes seperately 
pd.concat([df4,df5],join='outer',keys=['FirstDF','SecondDF'])
```

## Join() Method 

```python
#lets create 2 dataframes with different column name
ds1=pd.DataFrame({'A':[5,6,8,41,45,6],'B':['A','B','C','D','E','G']})
ds2=pd.DataFrame({'C':[15,16,18,13,5,14],'D':['Abc','Bef','Ceh','Dij','Elm','Gkp']})
ds3=pd.DataFrame({'G':[5,6,8],'H':['A','B','C']})
ds4=pd.DataFrame({'G':[15,46,18],'I':['Acd','Bef','Chi']})
display(ds1,ds2,ds3,ds4)
```

```python
a=ds1.join(ds2)
b=ds2.join(ds1)
display(a,b)
```

```python
#operation on dataframes of different indexes
a=ds3.join(ds1)#By default how='inner'
b=ds3.join(ds1,how='outer')
display(a,b)
```

```python
#when you have same columnname in both dataframes
ds3.join(ds4,lsuffix='1')
```

## Append() Function

```python
info1 = pd.DataFrame({"x":[25,15,12,19],   
                    "y":[47, 24, 17, 29]})     

info2 = pd.DataFrame({"x":[25, 15, 12],   
                    "y":[47, 24, 17],    
                    "z":[38, 12, 45]})   

info1.append(info2)
```

```python
info1.append(info2, ignore_index=True,sort=True)
```
