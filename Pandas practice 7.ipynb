# Some Other Important Functions 

## pivot_table()

```python
import pandas as pd
```

```python
data = pd.read_csv('D:\\Datasets\\Test Data\\Employee_Dataset_3.csv')
data
```

```python
#Use pivot_table with particular column
data.pivot_table(index='Department')
```

```python
#Use pivot_table with Multiple column
data.pivot_table(index='Department',columns=['Marital Status','Gender'])
```

```python
#functiion by defualt takes mean value. use aggfunc to to change it
data.pivot_table(index='Department',aggfunc='max')
```

```python
# fill 0 value to empty spacecs
data.pivot_table(index='Department',columns=['Marital Status'],fill_value=0)
```

```python
data.pivot_table(index='Department',margins=True)
```

## melt() function

```python
#use melt function on current dataset
pd.melt(data)
```

```python
pd.melt(data,id_vars=['Monthly Income','Age'])
```

```python
#change default variable and value columns
pd.melt(data,id_vars=['Monthly Income'],value_vars=['Education'])
```

```python
pd.melt(data,id_vars=['Monthly Income'],value_vars=['Education'],var_name='Col_name')
```
